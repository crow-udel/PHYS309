### Questions


Recall that the speed of light in vacuum is 
\begin{equation}
c = \frac{1}{\sqrt{\epsilon_0 \mu_0}}, \; \; \; \; \; (1)
\end{equation}
where $\epsilon_0 = 8.85 \times 10^{-12}$ F/m (Farads per meter)$^{*}$ is the electric permittivity of free space and $\mu_0 = 1.26 \times 10^{-6}$ N/A$^2$ (Newtons per Ampere$^2$)$^{**}$ is the magnetic permeability of free space. The electric field created by a point charge $q$ is
\begin{equation}
E = \frac{1}{4 \pi \epsilon_0} \frac{q}{r^2}. \; \; \; \; (2)
\end{equation}
Suppose a charged particle is moving at some fraction of the speed of light in vacuum ($v_q = f c; \; 0 < f < 1$) as seen in reference frame $S$. Reference frame $S^{\prime}$ is co-moving with the point charge, so that $\frac{dx^{\prime}}{dt^{\prime}} = v^{\prime}_q = 0$ in frame $S^{\prime}$. To make the math easier, put $q$ at the origin of $S^{\prime}$.

To understand Einstein's motivation for developing the special theory of relativity, make the <strong>incorrect, erroneous, and wrong</strong> assumption that the Galilean velocity transformation holds true for all values of $f$:
\begin{equation}
\frac{dx^{\prime}}{dt^{\prime}} = \frac{dx}{dt} - v_q \; \; \; (3)
\end{equation}

</ol>

```python
# Plot here
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const

xprime = np.linspace(1e-2, 1e-1, 1000)
field =  (-1 * const.e) / (4 * const.pi * const.epsilon_0 * xprime**2)

plt.plot(xprime, field)

plt.xlabel("Distance r' (m)")
plt.ylabel("Electric Field E (N/C)")
plt.title("Electric Field of an Electron vs Distance")
```

<li>If we calculate $c^{\prime}$ using the Galilean transformation in Equation 3, is there any value of $f$ for which $E^{\prime} = E$ <strong>and</strong> $B^{\prime} = B$? Make a graph to illustrate your answer.</li>

Since E depends on epsilon0 and mu depends on mu0 and these ar coupled by c the only way that E=E' AND B=B' is if epsilon0 = epsilon0' AND mu0=mu0' which is when c=c' or the rest frame. I try to show it in the graphs, but the important thing to note is that you CANNOT hold both constant without restricting c to also be constant.

```python
# Plot here
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const

f = np.linspace(0, 0.99, 1000)
vq = f * const.c

rprime = 1e-2

cprime = const.c - vq
epsilonprime = 1 /(cprime**2 * const.mu_0)
muprime = 1 / (cprime**2 * const.epsilon_0)

eprimeholdmu = (-1 * const.e) / (4 * const.pi * epsilonprime * rprime**2)
bprimeholdmu = const.mu_0 / (2*1) * np.linspace(1,1,1000)

eprimeholdep = (-1 * const.e) / (4 * const.pi * const.epsilon_0 * rprime**2) * np.linspace(1,1,1000)
bprimeholdep = muprime / (2*1)

fig, ax = plt.subplots(2, 1, sharex=True)

ax[0].plot(f, eprimeholdmu, label="E' with Original Mu")
ax[0].plot(f, bprimeholdmu, label="B' with Original Mu")
ax[0].set_ylabel("Fields (N/C or T)")
ax[0].set_title("Holding Mu Constant")

ax[1].plot(f, eprimeholdep, label="E' with Original Epsilon")
ax[1].plot(f, bprimeholdep, label="B' with Original Epsilon")
ax[1].set_xlabel("Fraction of Speed of Light (f)")
ax[1].set_ylabel("Fields (N/C or T)")
ax[1].set_title("Holding Epsilon Constant")

plt.tight_layout()
plt.show()

```
